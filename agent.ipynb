```python
import os
import re
from groq import Groq
```

```python

os.environ["GROQ_API_KEY"] =
```

```python
client = Groq()
```

```python
system_prompt = """
You are a reasoning agent that can use tools.

You must follow this format strictly:

Thought: describe what you need to do
PAUSE
Action: tool_name: tool_input

When you receive an Observation, continue reasoning.

When you have the final result, reply with:

Answer: <final answer>

Available tools:
- get_planet_mass(name)
- calculate(expression)
"""
```

```python
def get_planet_mass(name: str) -> float:
    masses = {
        "Earth": 5.972e24,
        "Saturn": 5.683e26,
    }
    return masses.get(name, 0.0)
```

```python
def calculate(expression: str) -> float:
    # entorno matemático seguro
    allowed = {"__builtins__": {}}
    return eval(expression, allowed)
```

```python
TOOLS = {
    "get_planet_mass": get_planet_mass,
    "calculate": calculate,
}
```

```python
class Agent:
    def __init__(
        self,
        client,
        system_prompt,
        model="llama-3.1-8b-instant",
        temperature=0,
    ):
        self.client = client
        self.model = model
        self.temperature = temperature
        self.messages = [{"role": "system", "content": system_prompt}]

    def __call__(self, message: str):
        if message:
            self.messages.append({"role": "user", "content": message})
        response = self._execute()
        self.messages.append({"role": "assistant", "content": response})
        return response

    def _execute(self):
        completion = self.client.chat.completions.create(
            model=self.model,
            messages=self.messages,
            temperature=self.temperature,
        )
        return completion.choices[0].message.content
```

```python
def loop(query: str, max_iterations: int = 10):
    agent = Agent(
        client=client,
        system_prompt=system_prompt,
        model="llama-3.1-8b-instant",
    )

    next_prompt = query

    for step in range(max_iterations):
        print(f"\n--- STEP {step + 1} ---")
        result = agent(next_prompt)
        print(result)

        # detectar acción
        if "PAUSE" in result and "Action" in result:
            match = re.search(r"Action:\s*([a-z_]+)\s*:\s*(.+)", result, re.IGNORECASE)

            if not match:
                next_prompt = "Observation: Invalid action format"
                continue

            tool_name, tool_input = match.groups()

            if tool_name not in TOOLS:
                next_prompt = "Observation: Tool not found"
                continue

            try:
                observation = TOOLS[tool_name](tool_input)
                next_prompt = f"Observation: {observation}"
            except Exception as e:
                next_prompt = f"Observation: Tool error - {e}"

            print(next_prompt)
            continue

        if "Answer" in result:
            break
```

```python
loop(
    query="What is the mass of Earth plus the mass of Saturn and all of that times 2?"
)
```

```python
from langchain_groq import ChatGroq
from langchain.agents import AgentExecutor, create_react_agent
from langchain.tools import tool
from langchain.prompts import PromptTemplate
```

```python
llm = ChatGroq(
    model="llama-3.1-8b-instant",
    temperature=0,
)
```

```python
@tool
def get_planet_mass(name: str) -> float:
    """Returns the mass of a planet in kilograms"""
    masses = {
        "Earth": 5.972e24,
        "Saturn": 5.683e26,
    }
    return masses.get(name, 0.0)
```

```python
@tool
def calculate(expression: str) -> float:
    """Evaluates a math expression"""
    allowed = {"__builtins__": {}}
    return eval(expression, allowed)
```

```python
tools = [get_planet_mass, calculate]
```

```python
prompt = PromptTemplate.from_template(
"""
You are a reasoning agent that can use tools.

You have access to the following tools:
{tools}

Use the following format:

Thought: your reasoning
Action: the action to take, one of [{tool_names}]
Action Input: the input to the action
Observation: the result of the action
... (this Thought/Action/Observation can repeat)

When you have the final answer, respond with:

Final Answer: the answer

Begin!

Question: {input}
{agent_scratchpad}
"""
)
```

```python
agent = create_react_agent(
    llm=llm,
    tools=tools,
    prompt=prompt,
)
```

```python
agent_executor = AgentExecutor(
    agent=agent,
    tools=tools,
    verbose=True,
)
```

```python
agent_executor.invoke(
    {
        "input": "What is the mass of Earth plus the mass of Saturn and all of that times 2?"
    }
)
```
